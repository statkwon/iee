```python
from iee import Explainer
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target
feature_names = diabetes.feature_names
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)
```

```python
explainer = Explainer(model)
iee_values = explainer(X_test)
```

```python
fig, ax = plt.subplots()
ax.barh(feature_names, iee_values[0])
ax.set_xlabel("IEE Values")
ax.set_title("Local IEE Values")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(feature_names, iee_values.mean(axis=0))
ax.set_xlabel("IEE Values")
ax.set_title("Global IEE Values")
plt.show()
```
